```python
predictors = company_imputed.drop('default', axis =1)
response = company_imputed['default']
```

```python
predictors.shape
```

```python
# Split the data into train and test

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.feature_selection import RFE
```

```python
x_train,x_test,y_train,y_test = train_test_split(predictors, response, test_size=0.25,random_state=2, stratify= response)
```

```python
# For modeling we will use logistic regression with RFE (Recurssive Feature Elimination)

```

```python
logR = LogisticRegression()
# logR.fit(x_train,y_train)
```

```python
predictors.shape
```

```python
# n_features_to_select = 1/3 *43 = 14.33(approx 14) --- this is thumb rule to select n_features_to_select equals to 1/3 of total no of columns of independent variables(predicators)
```

```python
selector = RFE(estimator=logR, n_features_to_select=15, step=1)
```

```python
# step=1 means RFE will denote first most 15 significant features by 1.
```

```python
selector = selector.fit(x_train, y_train)
```

```python
selector.n_features_
```

```python
selector.ranking_
```

```python
# here we have given n_features_to_select = 15, it means that out of 43 features RFE will select 15 most significant features.
# We can see 1 appears 15 times. Start indexing from 0 and check at which index 1 appears, only that features is significant according to RFE
```

```python
# we are going to get features(column name) which appears at index on which 1 appears.

df = pd.DataFrame({'feature' : scaled_predictor.columns, 'Rank': selector.ranking_})
df[df['Rank']==1]
```

```python
# correlation analysis
sns.heatmap(company_imputed.corr())
```

```python
# This is another approach to get which features are most significant but RFE is more reliable

df1 = company_imputed.corr()['default'].iloc[1:].to_frame().reset_index().rename(columns={'index':'feature', 
                                                                                         'default':'corr_coeff'})
df1
```

# Validating the model on train test set

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
pred_train = selector.predict(x_train)
pred_test = selector.predict(x_test)
```

```python
print(confusion_matrix(y_train, pred_train))
print("*****************"*10)
print(confusion_matrix(y_test, pred_test))
```

```python
print("Training Accuracy", accuracy_score(y_train, pred_train))
print("**************"*10)
print("Test Accuracy", accuracy_score(y_test, pred_test))
```

```python
print(classification_report(y_train, pred_train))
print("**************"*10)
print(classification_report(y_test, pred_test))
```

```python
response.value_counts(normalize=True)
```

# We see poor recall score for both train and test

## Since, only 7% of the total data had defaults (-1), we will now try to balance the data before fitting the model¶

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=33)
x_sm, y_sm = sm.fit_resample(x_train, y_train)
```

```python
y_train.value_counts(normalize = True)
```

```python
y_sm.value_counts(normalize = True)
```

```python
selector_smoter = selector.fit(x_sm, y_sm)
```

```python
selector_smoter.n_features_
```

```python
pred_train_smoter = selector_smoter.predict(x_sm)
pred_test_smoter = selector_smoter.predict(x_test)
```

```python
print("Training Accuracy :", accuracy_score(y_sm, pred_train_smoter))
print("**************"*10)
print("Test Accuracy :", accuracy_score(y_test, pred_test_smoter))
```

```python
print(classification_report(y_sm, pred_train_smoter))
print("**************"*10)
print(classification_report(y_test, pred_test_smoter))
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rForest = RandomForestClassifier()
```

```python
select_rf = RFE(estimator=rForest, n_features_to_select=15, step=1)
```

```python
select_rf = select_rf.fit(x_sm, y_sm)
```

```python
pred_train_rf_smote= select_rf.predict(x_sm)
pred_test_rf_smote = select_rf.predict(x_test)
```

```python
print("Training Accuracy", accuracy_score(y_sm, pred_train_rf_smote))
print("**************"*10)
print("Test Accuracy", accuracy_score(y_test, pred_test_rf_smote))
```

```python
# here in random forest with smote we are getting overfitting problem, so to reduce overfitting we have to  use cross validation method.
```

```python
# using cross validation approach
from sklearn.model_selection import cross_val_score
training = cross_val_score(select_rf, x_train, y_train, cv=10)
print("All 10 accuracy", training)
print("Avg Training Acc", training.mean())
print("Max Training Acc", training.max())
```
